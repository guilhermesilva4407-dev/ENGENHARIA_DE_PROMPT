```python
# Importa a biblioteca 'requests' para fazer requisições HTTP a APIs.
# Se você estiver em um ambiente Colab novo ou diferente, pode precisar instalá-la.
# Descomente e execute a linha abaixo uma vez:
!pip install requests

import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um CEP.

    Args:
        cep (str): O número do CEP a ser consultado (apenas dígitos).

    Returns:
        dict or None: Um dicionário com os dados do endereço formatados, ou None se houver erro.
    """
    # Limpa o CEP, removendo caracteres não numéricos
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem 8 dígitos, que é o formato esperado pela ViaCEP
    if len(cep) != 8:
        print(f"Erro: O CEP '{cep}' é inválido. Ele deve conter 8 dígitos numéricos.")
        return None

    # Monta a URL da API ViaCEP
    url = f"https://viacep.com.br/ws/{cep}/json/"

    print(f"Consultando CEP: {cep} na URL: {url}")

    try:
        # Faz a requisição HTTP GET para a API
        # timeout=10 define um tempo limite de 10 segundos para a requisição.
        response = requests.get(url, timeout=10)
        response.raise_for_status() # Lança um HTTPError para respostas de status de erro (4xx ou 5xx)

        # Converte a resposta JSON em um dicionário Python
        dados_cep = response.json()

        # A API ViaCEP retorna um dicionário com {'erro': True} se o CEP não for encontrado
        if 'erro' in dados_cep and dados_cep['erro']:
            print(f"Erro: CEP '{cep}' não encontrado.")
            return None
        else:
            # Formata e retorna os dados do endereço
            endereco_formatado = {
                "CEP": dados_cep.get('cep', 'N/A'),
                "Logradouro": dados_cep.get('logradouro', 'N/A'),
                "Complemento": dados_cep.get('complemento', ''),
                "Bairro": dados_cep.get('bairro', 'N/A'),
                "Localidade": dados_cep.get('localidade', 'N/A'),
                "UF": dados_cep.get('uf', 'N/A'),
                "DDD": dados_cep.get('ddd', 'N/A')
            }
            return endereco_formatado

    except requests.exceptions.HTTPError as http_err:
        # Erros de resposta HTTP (status 4xx, 5xx)
        print(f"Erro HTTP ao consultar o CEP '{cep}': {http_err}")
    except requests.exceptions.ConnectionError as conn_err:
        # Erros de conexão (DNS, rede, etc.)
        print(f"Erro de Conexão ao consultar o CEP '{cep}': {conn_err}")
    except requests.exceptions.Timeout as timeout_err:
        # Erros de tempo limite de requisição
        print(f"Erro de Timeout ao consultar o CEP '{cep}': {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        # Outros erros da biblioteca requests
        print(f"Erro de Requisição inesperado ao consultar o CEP '{cep}': {req_err}")
    except ValueError as json_err:
        # Erro ao decodificar JSON (resposta não é um JSON válido)
        print(f"Erro ao decodificar a resposta JSON para o CEP '{cep}': {json_err}. Resposta: {response.text}")
    except Exception as e:
        # Qualquer outro erro inesperado
        print(f"Ocorreu um erro inesperado ao consultar o CEP '{cep}': {e}")

    return None

# --- Exemplo de Uso Interativo ---

print("\n--- Consulta de CEP Interativa ---")
# Solicita ao usuário que digite um CEP
cep_digitado = input("Por favor, digite um CEP para consultar (ex: 01001-000): ")

# Chama a função para consultar o CEP digitado pelo usuário
endereco_usuario = consultar_cep(cep_digitado)

# Exibe as informações do endereço se a consulta for bem-sucedida
if endereco_usuario:
    print("Informações do Endereço:")
    for chave, valor in endereco_usuario.items():
        print(f"  {chave}: {valor}")
else:
    print(f"Não foi possível obter informações para o CEP '{cep_digitado}'. Por favor, tente novamente com um CEP válido.")
```
